```python
import pandas as pd
from scipy import stats

# --- Comprehensive P-value Testing for 2D and 3D Features ---

def perform_4_group_analysis(feature_name, dataframe):
    """
    Performs a 4-group statistical comparison for a given feature.
    - Kruskal-Wallis for overall difference.
    - Mann-Whitney U for pairwise differences with Bonferroni correction.
    """
    print("\n" + "="*60)
    print(f"Analysis for Feature: '{feature_name}'")
    print("="*60)

    # 1. Filter data to create the four specific subgroups
    df_base = dataframe[
        (dataframe['turning_angle'] == '180_degrees') &
        (dataframe['type_of_turn'] == 'pivot_turn')
    ].copy()

    group_C = df_base[df_base['PD_or_C'] == 'C']
    group_PD_off = df_base[df_base['On_or_Off_medication'] == 'Off medication']
    group_PD_on = df_base[(df_base['On_or_Off_medication'] == 'On medication') & (df_base['DBS_state'] == '-')]
    group_PD_dbs = df_base[(df_base['On_or_Off_medication'] == 'On medication') & (df_base['DBS_state'] == 'On DBS')]

    # Check if we have data in all groups to perform the test
    if any(g.empty for g in [group_C, group_PD_off, group_PD_on, group_PD_dbs]):
        print("One or more subgroups are empty for this feature. Skipping test.")
        return

    # 2. Kruskal-Wallis test
    try:
        h_statistic, p_value_kruskal = stats.kruskal(
            group_C[feature_name].dropna(),
            group_PD_off[feature_name].dropna(),
            group_PD_on[feature_name].dropna(),
            group_PD_dbs[feature_name].dropna()
        )
    except ValueError:
        print("Could not perform Kruskal-Wallis test (likely not enough data in one group).")
        return

    print(f"Overall Kruskal-Wallis Test -> P-value: {p_value_kruskal:.4f}")

    if p_value_kruskal < 0.05:
        print("--> Overall difference is significant. Performing pairwise tests...")

        # 3. Post-Hoc Pairwise Comparisons
        corrected_alpha = 0.05 / 6 # Bonferroni correction for 6 pairs

        comparisons = {
            "Control vs. PD Off-Med": (group_C, group_PD_off),
            "Control vs. PD On-Med": (group_C, group_PD_on),
            "Control vs. PD On-Med+DBS": (group_C, group_PD_dbs),
            "PD Off-Med vs. PD On-Med": (group_PD_off, group_PD_on),
            "PD Off-Med vs. PD On-Med+DBS": (group_PD_off, group_PD_dbs),
            "PD On-Med vs. PD On-Med+DBS": (group_PD_on, group_PD_dbs),
        }

        for name, (g1, g2) in comparisons.items():
            if g1.empty or g2.empty: continue
            stat, p_val = stats.mannwhitneyu(g1[feature_name].dropna(), g2[feature_name].dropna(), alternative='two-sided')
            significance = "Significant" if p_val < corrected_alpha else "Not Significant"
            print(f"  - {name:<28}: P-value = {p_val:<7.4f} ({significance})")
    else:
        print("--> No significant overall difference among the groups.")

# Define the lists of features to test
kinematic_features_2d = [
    'duration_sec', 'pelvis_path_length', 'pelvis_peak_speed',
    'pelvis_avg_accel_mag', 'pelvis_avg_jerk_mag', 'trunk_total_rotation_deg',
    'trunk_avg_ang_vel', 'trunk_peak_ang_vel'
]
kinematic_features_3d = [
    'duration_sec', 'pelvis_path_length_3d', 'pelvis_peak_speed_3d',
    'pelvis_avg_accel_mag_3d', 'pelvis_avg_jerk_mag_3d',
    'pelvis_vertical_displacement', 'pelvis_avg_vertical_speed',
    'mean_body_height', 'trunk_total_rotation_deg',
    'trunk_avg_ang_vel', 'trunk_peak_ang_vel'
]

# --- Run Analysis for 2D Data ---
print("\n\n###########################################")
print("###      ANALYSIS ON 2D FEATURES      ###")
print("###########################################")
if 'df_2d_features' in locals():
    for feature in kinematic_features_2d:
        perform_4_group_analysis(feature, df_2d_features)
else:
    print("2D features DataFrame not found.")

# --- Run Analysis for 3D Data ---
print("\n\n###########################################")
print("###      ANALYSIS ON 3D FEATURES      ###")
print("###########################################")
if 'df_3d_features' in locals():
    for feature in kinematic_features_3d:
        perform_4_group_analysis(feature, df_3d_features)
else:
    print("3D features DataFrame not found.")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.utils import resample

# --- Binary Model: PD On-Med (No DBS) vs. PD On-Med (DBS On) ---

if 'df_2d_features' in locals():
    # 1. Filter for the specific cohort
    df_base = df_3d_features[
        (df_3d_features['turning_angle'] == '180_degrees') &
        (df_3d_features['type_of_turn'] == 'pivot_turn') &
        (df_3d_features['On_or_Off_medication'] == 'On medication') &
        (df_3d_features['DBS_state'].isin(['-', 'On DBS']))
    ].copy()

    print("--- Preparing data for the DBS effect model ---")
    print("Class distribution for this task:")
    print(df_base['DBS_state'].value_counts())

    # 2. Define features (kinematic only) and target
    kinematic_features_3d = [
        'duration_sec', 'pelvis_path_length_3d', 'pelvis_displacement_3d', 'pelvis_avg_speed_3d',
        'pelvis_peak_speed_3d', 'pelvis_avg_accel_mag_3d', 'pelvis_peak_accel_mag_3d',
        'pelvis_avg_jerk_mag_3d', 'pelvis_vertical_displacement', 'pelvis_avg_vertical_speed',
        'mean_body_height', 'trunk_total_rotation_deg', 'trunk_avg_ang_vel', 'trunk_peak_ang_vel'
    ]

    X = df_base[kinematic_features_3d]
    y = df_base['DBS_state']

    # 3. Train-Test Split
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.3, random_state=42, stratify=y
    )

    # 4. Apply Undersampling to the Training Data
    df_train = pd.concat([X_train, y_train], axis=1)
    df_no_dbs_train = df_train[df_train['DBS_state'] == '-']
    df_dbs_on_train = df_train[df_train['DBS_state'] == 'On DBS']

    # Undersample the majority class ('-')
    df_majority_downsampled = resample(df_no_dbs_train,
                                     replace=False,
                                     n_samples=len(df_dbs_on_train),
                                     random_state=42)

    df_train_downsampled = pd.concat([df_majority_downsampled, df_dbs_on_train])

    print("\nTraining set class distribution after undersampling:")
    print(df_train_downsampled['DBS_state'].value_counts())

    X_train_under = df_train_downsampled.drop('DBS_state', axis=1)
    y_train_under = df_train_downsampled['DBS_state']

    # 5. Scale the features
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train_under)
    X_test_scaled = scaler.transform(X_test)

    # 6. Train and Evaluate the model
    model = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
    model.fit(X_train_scaled, y_train_under)
    y_pred = model.predict(X_test_scaled)
    accuracy = accuracy_score(y_test, y_pred)

    print("\n" + "="*60)
    print("     DBS Model Evaluation: No DBS vs. On DBS")
    print("="*60)
    print(f"Overall Model Accuracy: {accuracy:.4f} ({accuracy:.2%})\n")
    print("Classification Report:")
    print(classification_report(y_test, y_pred))

else:
    print("DataFrame 'df_3d_features' not found.")
```
